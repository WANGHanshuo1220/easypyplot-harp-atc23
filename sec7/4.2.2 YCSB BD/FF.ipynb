```python
''' Required packages. Use python3 -m pip install xxx for all the packages. '''
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
from matplotlib import pyplot as plt
import easypyplot
```

```python
''' Plot parameters. Change only if you want to resize the graph, etc. '''
fig_bar_width = 0.75
fig_dims = (35, 10)
fig_label_fontsize = 35
xy_label_size = 23
ft = fm.FontProperties('Times New Roman')
ft.set_size(fig_label_fontsize)

fig_name = 'graph_sensitivity'
```

```python
workload_list = ['18', '24', '30', '36']
fig_scheme_list = ['Baseline Read Access', 'Baseline Write Access',
                   'Harp Read Access', 'Harp Write Access']
workload_dict = {item: item for item in workload_list}
fig_scheme_dict = {item: item for item in fig_scheme_list}

fig_title = 'YCSB Breakdown'
fig_name = 'GC breakdown'
fig_path = '../../4.2.2/'
fig_ylim = (0, 1.2)
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1]+0.01, 0.2))
fig_bar_colors = [easypyplot.color.COLOR_SET[i] for i in [4, 0]]
design_colors = fig_bar_colors
baseline_colors = easypyplot.color.color_scale(design_colors[0], 2, 0.5, 0.75)[::-1]
igc_colors = easypyplot.color.color_scale(design_colors[1], 2, 0.5, 0.75)[::-1]
design_len = 2
assert len(fig_bar_colors) == design_len
```

```python
''' Load CCEH data '''
csv_path = 'CCEH.csv'

# pprint(fig_yticks)
csv_hd = pd.read_csv(csv_path, sep=',')
fig_xlabels = []
baseline_xticks = []
fig_data_CCEH = []
xtick = -1
for idx, row in csv_hd.iterrows():
    workload_fullname = str(int(row[0]))
    if any(workload == workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in fig_scheme_dict):
                row_value.append(row[col])
        fig_data_CCEH.append(row_value)
        fig_xlabels.append(workload_fullname)
        xtick += 2
        baseline_xticks.append(xtick)
assert len(fig_data_CCEH) == len(fig_xlabels) == len(baseline_xticks)
igc_xticks = [item + fig_bar_width for item in baseline_xticks]

pprint(fig_data_CCEH)

''' Layout hit rate '''
layout_scheme_list = ['Baseline GC', 'Harp GC']
layout_scheme_dict = {item: item for item in layout_scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
layout_xlabels = []
layout_xticks = []
layout_data_CCEH = []
xtick = -1
for idx, row in csv_hd.iterrows():
    workload_fullname = str(int(row[0]))
    if any(workload == workload_fullname for workload in workload_list):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in layout_scheme_dict):
                row_value.append(row[col])
        layout_data_CCEH.append(row_value)
        layout_xlabels.append(workload_fullname)
        xtick += 1
        layout_xticks.append(xtick)
assert len(layout_data_CCEH) == len(layout_xlabels) == len(layout_xticks)

pprint(layout_data_CCEH)
```

```python
pp, fig = easypyplot.pdf.plot_setup(fig_name, fig_dims)

''' plot CCEH '''
fig_ax = fig.add_subplot(1 ,4 ,1)
fig_ax2 = fig_ax.twinx()
easypyplot.format.turn_off_box(fig_ax, twinx_axes=fig_ax2)
fig_data_nparr = np.array(fig_data_CCEH)
baseline_acc_data = fig_data_nparr[:, :2]
igc_acc_data = fig_data_nparr[:, 2:4]
fig_h = []
fig_h += easypyplot.barchart.draw(
    fig_ax, baseline_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=baseline_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=baseline_colors)
fig_h += easypyplot.barchart.draw(
    fig_ax, igc_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=igc_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=igc_colors)
fig_h2 = []
# easypyplot.format.turn_off_box(ax2)
layout_data_nparr = np.array(layout_data_CCEH)
layout_xticks = [baseline_xticks, igc_xticks]
for group_id in range(design_len):
    print(layout_data_nparr[:, 0])
    fig_h2 += fig_ax2.plot(layout_xticks[group_id], layout_data_nparr[:, group_id],
                marker='*',
                linewidth = '1.0',
                color = '#000000',
                # color = baseline_colors[0],
                markersize = '7.0',
                # prop = ft
                )
''' x axis '''
fig_ax.set_xlim([fig_ax.get_xticks()[0] - 1, fig_ax.get_xticks()[-1] + 1])
fig_ax.set_xticks(baseline_xticks)
fig_ax.set_xlim([0, len(workload_list) * design_len + 1])
fig_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# fig_ax.tick_params(direction='in') # [in, out, inout]
fig_ax.set_xlabel('CCEH', fontproperties=ft)
''' y axis '''
fig_ax.yaxis.grid(True)
fig_ax.set_ylim(fig_ylim)
fig_ax.set_yticks(fig_yticks)
fig_ax.set_ylabel('Access times (million)', multialignment='center', fontproperties=ft)
layout_ylim = [0., 15]
fig_ax2.set_ylim(layout_ylim)
fig_ax2.set_yticks(list(np.arange(layout_ylim[0], layout_ylim[1]+0.01, 3)))
# fig_ax.set_yticklabels(fontsize=fig_label_fontsize)
# fig_ax2.set_yticklabels(fontsize=fig_label_fontsize)
# fig_ax2.set_ylabel('GC Latency', multialignment='center', fontproperties=ft)
fig_ax.tick_params(labelsize=xy_label_size)
fig_ax2.tick_params(labelsize=xy_label_size)
easypyplot.format.resize_ax_box(fig_ax, hratio=0.8)

''' plot CCEH-P '''
fig_ax = fig.add_subplot(1 ,4 ,2)
fig_ax2 = fig_ax.twinx()
easypyplot.format.turn_off_box(fig_ax, twinx_axes=fig_ax2)
fig_data_nparr = np.array(fig_data_CCEH_P)
baseline_acc_data = fig_data_nparr[:, :2]
igc_acc_data = fig_data_nparr[:, 2:4]
# fig_h = []
fig_h += easypyplot.barchart.draw(
    fig_ax, baseline_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=baseline_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=baseline_colors)
fig_h += easypyplot.barchart.draw(
    fig_ax, igc_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=igc_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=igc_colors)
# fig_h2 = []
# easypyplot.format.turn_off_box(ax2)
layout_data_nparr = np.array(layout_data_CCEH_P)
layout_xticks = [baseline_xticks, igc_xticks]
for group_id in range(design_len):
    print(layout_data_nparr[:, 0])
    fig_h2 += fig_ax2.plot(layout_xticks[group_id], layout_data_nparr[:, group_id],
                marker='*',
                linewidth = '1.0',
                color = '#000000',
                # color = baseline_colors[0],
                markersize = '7.0',
                # prop = ft
                )
''' x axis '''
fig_ax.set_xlim([fig_ax.get_xticks()[0] - 1, fig_ax.get_xticks()[-1] + 1])
fig_ax.set_xticks(baseline_xticks)
fig_ax.set_xlim([0, len(workload_list) * design_len + 1])
fig_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# fig_ax.tick_params(direction='in') # [in, out, inout]
fig_ax.set_xlabel('CCEH-P', fontproperties=ft)
''' y axis '''
fig_ax.yaxis.grid(True)
fig_ax.set_ylim(fig_ylim)
fig_ax.set_yticks(fig_yticks)
# fig_ax.set_ylabel('Access Breakdown', multialignment='center', fontproperties=ft)
layout_ylim = [0., 15]
fig_ax2.set_ylim(layout_ylim)
fig_ax2.set_yticks(list(np.arange(layout_ylim[0], layout_ylim[1]+0.01, 3)))
# fig_ax2.set_ylabel('GC Latency', multialignment='center', fontproperties=ft)
fig_ax.tick_params(labelsize=xy_label_size)
fig_ax2.tick_params(labelsize=xy_label_size)
easypyplot.format.resize_ax_box(fig_ax, hratio=0.8)

''' plot FF '''
fig_ax = fig.add_subplot(1 ,4 ,3)
fig_ax2 = fig_ax.twinx()
easypyplot.format.turn_off_box(fig_ax, twinx_axes=fig_ax2)
fig_data_nparr = np.array(fig_data_FF)
baseline_acc_data = fig_data_nparr[:, :2]
igc_acc_data = fig_data_nparr[:, 2:4]
# fig_h = []
fig_h += easypyplot.barchart.draw(
    fig_ax, baseline_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=baseline_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=baseline_colors)
fig_h += easypyplot.barchart.draw(
    fig_ax, igc_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=igc_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=igc_colors)
# fig_h2 = []
# easypyplot.format.turn_off_box(ax2)
layout_data_nparr = np.array(layout_data_FF)
layout_xticks = [baseline_xticks, igc_xticks]
for group_id in range(design_len):
    print(layout_data_nparr[:, 0])
    fig_h2 += fig_ax2.plot(layout_xticks[group_id], layout_data_nparr[:, group_id],
                marker='*',
                linewidth = '1.0',
                color = '#000000',
                # color = baseline_colors[0],
                markersize = '7.0',
                # prop = ft
                )
''' x axis '''
fig_ax.set_xlim([fig_ax.get_xticks()[0] - 1, fig_ax.get_xticks()[-1] + 1])
fig_ax.set_xticks(baseline_xticks)
fig_ax.set_xlim([0, len(workload_list) * design_len + 1])
fig_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# fig_ax.tick_params(direction='in') # [in, out, inout]
fig_ax.set_xlabel('FF', fontproperties=ft)
''' y axis '''
fig_ax.yaxis.grid(True)
fig_ax.set_ylim(fig_ylim)
fig_ax.set_yticks(fig_yticks)
# fig_ax.set_ylabel('Access Breakdown', multialignment='center', fontproperties=ft)
layout_ylim = [0., 15]
fig_ax2.set_ylim(layout_ylim)
fig_ax2.set_yticks(list(np.arange(layout_ylim[0], layout_ylim[1]+0.01, 3)))
# fig_ax2.set_ylabel('GC Latency', multialignment='center', fontproperties=ft)
fig_ax.tick_params(labelsize=xy_label_size)
fig_ax2.tick_params(labelsize=xy_label_size)
easypyplot.format.resize_ax_box(fig_ax, hratio=0.8)

''' plot MT '''
fig_ax = fig.add_subplot(1 ,4 ,4)
fig_ax2 = fig_ax.twinx()
easypyplot.format.turn_off_box(fig_ax, twinx_axes=fig_ax2)
fig_data_nparr = np.array(fig_data_MT)
baseline_acc_data = fig_data_nparr[:, :2]
igc_acc_data = fig_data_nparr[:, 2:4]
# fig_h = []
fig_h += easypyplot.barchart.draw(
    fig_ax, baseline_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=baseline_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=baseline_colors)
fig_h += easypyplot.barchart.draw(
    fig_ax, igc_acc_data,
    width=fig_bar_width,
    breakdown=True,
    group_names=fig_xlabels,
    xticks=igc_xticks,
    xticklabelfontsize=fig_label_fontsize,
    xticklabelrotation=0,
    # prop=ft,
    colors=igc_colors)
# fig_h2 = []
# easypyplot.format.turn_off_box(fig)
layout_data_nparr = np.array(layout_data_MT)
layout_xticks = [baseline_xticks, igc_xticks]
for group_id in range(design_len):
    print(layout_data_nparr[:, 0])
    fig_h2 += fig_ax2.plot(layout_xticks[group_id], layout_data_nparr[:, group_id],
                marker='*',
                linewidth = '1.0',
                color = '#000000',
                # color = baseline_colors[0],
                markersize = '7.0',
                # prop = ft
                )
''' x axis '''
fig_ax.set_xlim([fig_ax.get_xticks()[0] - 1, fig_ax.get_xticks()[-1] + 1])
fig_ax.set_xticks(baseline_xticks)
fig_ax.set_xlim([0, len(workload_list) * design_len + 1])
fig_ax.set_xticklabels(fig_xlabels, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax.xaxis.set_ticks_position('none') # Comment this and the next line for visible ticks
# fig_ax.tick_params(direction='in') # [in, out, inout]
fig_ax.set_xlabel('Masstree', fontproperties=ft)
''' y axis '''
fig_ax.yaxis.grid(True)
fig_ax.set_ylim(fig_ylim)
fig_ax.set_yticks(fig_yticks)
# fig_ax.set_ylabel('Access Breakdown', multialignment='center', fontproperties=ft)
layout_ylim = [0., 15]
fig_ax2.set_ylim(layout_ylim)
fig_ax2.set_yticks(list(np.arange(layout_ylim[0], layout_ylim[1]+0.01, 3)))
fig_ax2.set_ylabel('GC times (thousand)', multialignment='center', fontproperties=ft)

''' legend '''
legend = [fig_scheme_dict[item] for item in fig_scheme_dict]
fig.legend(fig_h, legend, loc='upper center', bbox_to_anchor=(0.5, 1),
        ncol=4, frameon=False, prop=ft, columnspacing=3, labelspacing=0)

# ''' last resize '''
# plt.subplots_adjust(right=1.2)
# fig.subplots_adjust(hspace=1)
# plt.tight_layout()
fig_ax.tick_params(labelsize=xy_label_size)
fig_ax2.tick_params(labelsize=xy_label_size)
easypyplot.format.resize_ax_box(fig_ax, hratio=0.8)
```
