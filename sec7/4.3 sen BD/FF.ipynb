```python
''' Required packages. Use python3 -m pip install xxx for all the packages. '''
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
from matplotlib import pyplot as plt
import easypyplot
```

```python
''' Plot parameters. Change only if you want to resize the graph, etc. '''
fig_bar_width = 0.75
fig_dims = (10, 2)
fig_label_fontsize = 12
xy_label_size = 10
ft = fm.FontProperties('Times New Roman')
ft.set_size(fig_label_fontsize)

fig_name = 'graph_sensitivity'
```

```python
fig_scheme_list = ['Baseline Read Access', 'Baseline Write Access',
                   'Harp Read Access', 'Harp Write Access']
fig_scheme_dict = {item: item for item in fig_scheme_list}

fig_title = 'Sensitivity'
fig_name = '4.3 Sensitivity breakdown'
fig_path = '../../4.3/'
fig_ylim = (0, 2)
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1]+0.01, 0.4))
fig_bar_colors = [easypyplot.color.COLOR_SET[i] for i in [4, 0]]
design_colors = fig_bar_colors
baseline_colors = easypyplot.color.color_scale(design_colors[0], 2, 0.5, 0.75)[::-1]
igc_colors = easypyplot.color.color_scale(design_colors[1], 2, 0.5, 0.75)[::-1]
design_len = 2
assert len(fig_bar_colors) == design_len
```

```python
''' Load wirte ratio data '''
csv_path = 'write_ratio.csv'
workload_list_WR = ['0', '20', '40', '60', '80', '100']
workload_dict = {item: item for item in workload_list_WR}

# pprint(fig_yticks)
csv_hd = pd.read_csv(csv_path, sep=',')
fig_xlabels_WR = []
baseline_xticks_WR = []
fig_data_WR = []
xtick_WR = -1
for idx, row in csv_hd.iterrows():
    workload_fullname = str(int(row[0]))
    if any(workload == workload_fullname for workload in workload_list_WR):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in fig_scheme_dict):
                row_value.append(row[col])
        fig_data_WR.append(row_value)
        fig_xlabels_WR.append(workload_fullname)
        xtick_WR += 2
        baseline_xticks_WR.append(xtick_WR)
assert len(fig_data_WR) == len(fig_xlabels_WR) == len(baseline_xticks_WR)
igc_xticks_WR = [item + fig_bar_width for item in baseline_xticks_WR]

pprint(fig_data_WR)

''' Layout hit rate '''
layout_scheme_list = ['Baseline GC', 'Harp GC']
layout_scheme_dict = {item: item for item in layout_scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
layout_xlabels_WR = []
layout_xticks_WR = []
layout_data_WR = []
xtick_WR = -1
for idx, row in csv_hd.iterrows():
    workload_fullname = str(int(row[0]))
    if any(workload == workload_fullname for workload in workload_list_WR):
        row_value = []
        for col in row.keys():
            if any(scheme in col for scheme in layout_scheme_dict):
                row_value.append(row[col])
        layout_data_WR.append(row_value)
        layout_xlabels_WR.append(workload_fullname)
        xtick_WR += 1
        layout_xticks_WR.append(xtick_WR)
assert len(layout_data_WR) == len(layout_xlabels_WR) == len(layout_xticks_WR)

pprint(layout_data_WR)
```
