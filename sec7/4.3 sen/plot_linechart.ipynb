```python
''' Required packages. Use python3 -m pip install xxx for all the packages. '''
from pprint import pprint
import pandas as pd
import numpy as np
import matplotlib.font_manager as fm
from matplotlib import pyplot as plt
import easypyplot
# Automatically update if external package code is modified
%load_ext autoreload
%autoreload 2
```

```python
''' Plot knobs. Change only if you want to resize the graph, etc. '''
fig_dims = (10, 2)
fig_label_fontsize = 14
xy_label_size = 8

ft = fm.FontProperties('Times New Roman')
ft.set_size(fig_label_fontsize)

''' [TODO] Figure properties. Specify the csv file and corresponding labels. '''
workload_list = ['Harp', 'Pacman'] # TODO: list of all workloads you have used. Each column corresponds with one workload
workload_dict = {item: item for item in workload_list} # Change the mapping if you want a different figure label text instead of the csv column name

fig_title = 'Throughput (Mops)' # TODO: figure title, shown as y-axis label
fig_path = '../../4.3/'
fig_name = '4.3 Sensitivity throughput' # TODO: figure name, used as pdf name
# fig_ylim = (0, 50) # TODO: y-axis range. Align with your data
# fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1], 10.0))
fig_colors = [easypyplot.color.COLOR_SET[i] for i in [1, 0]] # TODO: choose enough colors for schemes
FULL_MARKER_SET = ['o', '^', 'v', 's', 'D']
fig_marker = [FULL_MARKER_SET[i] for i in [0, 1]] # TODO: choose enough markers for schemes
assert len(fig_colors) == len(workload_dict)
assert len(fig_marker) == len(workload_dict)
```

```python
''' Load write ratio CCEH. '''
csv_path = 'write_ratio.csv' # TODO: path to the result table, in csv format
scheme_list = ['0', '20', '40', '60', '80', '100'] # TODO: list of all schemes in the breakdown. Each row corresponds with one scheme
scheme_dict = {item: item for item in scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
# TODO assert workload_name
# TODO assert scheme_name
fig_xlabels_WR = []
fig_xticks_WR = []
fig_data_WR = []
xtick = -1
for idx, row in csv_hd.iterrows():
    scheme_fullname = str(int(row[0]))
    if any(scheme in scheme_fullname for scheme in scheme_list):
        row_value = []
        for col in row.keys():
            if any(workload in col for workload in workload_dict):
                row_value.append(row[col])
        fig_data_WR.append(row_value)
        fig_xlabels_WR.append(scheme_fullname)
        xtick += 1
    fig_xticks_WR.append(xtick)
fig_data_WR = np.transpose(fig_data_WR)
assert fig_data_WR.shape[1] == len(fig_xlabels_WR) == len(fig_xticks_WR)

pprint(fig_data_WR)

''' Load data util. '''
csv_path = 'util.csv' # TODO: path to the result table, in csv format
scheme_list = ['50', '60', '70', '80'] # TODO: list of all schemes in the breakdown. Each row corresponds with one scheme
scheme_dict = {item: item for item in scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
# TODO assert workload_name
# TODO assert scheme_name
fig_xlabels_U = []
fig_xticks_U = []
fig_data_U = []
xtick = -1
for idx, row in csv_hd.iterrows():
    scheme_fullname = str(int(row[0]))
    if any(scheme in scheme_fullname for scheme in scheme_list):
        row_value = []
        for col in row.keys():
            if any(workload in col for workload in workload_dict):
                row_value.append(row[col])
        fig_data_U.append(row_value)
        fig_xlabels_U.append(scheme_fullname)
        xtick += 1
    fig_xticks_U.append(xtick)
fig_data_U = np.transpose(fig_data_U)
assert fig_data_U.shape[1] == len(fig_xlabels_U) == len(fig_xticks_U)

pprint(fig_data_U)

''' Load data kv size. '''
csv_path = 'kv_size.csv' # TODO: path to the result table, in csv format
scheme_list = ['32', '64', '128', '256', '512'] # TODO: list of all schemes in the breakdown. Each row corresponds with one scheme
scheme_dict = {item: item for item in scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
# TODO assert workload_name
# TODO assert scheme_name
fig_xlabels_KVS = []
fig_xticks_KVS = []
fig_data_KVS = []
xtick = -1
for idx, row in csv_hd.iterrows():
    scheme_fullname = str(int(row[0]))
    if any(scheme in scheme_fullname for scheme in scheme_list):
        row_value = []
        for col in row.keys():
            if any(workload in col for workload in workload_dict):
                row_value.append(row[col])
        fig_data_KVS.append(row_value)
        fig_xlabels_KVS.append(scheme_fullname)
        xtick += 1
    fig_xticks_KVS.append(xtick)
fig_data_KVS = np.transpose(fig_data_KVS)
assert fig_data_KVS.shape[1] == len(fig_xlabels_KVS) == len(fig_xticks_KVS)

pprint(fig_data_KVS)

''' Load data key space. '''
csv_path = 'key_space.csv' # TODO: path to the result table, in csv format
scheme_list = ['100', '200', '400', '800'] # TODO: list of all schemes in the breakdown. Each row corresponds with one scheme
scheme_dict = {item: item for item in scheme_list}
csv_hd = pd.read_csv(csv_path, sep=',')
# TODO assert workload_name
# TODO assert scheme_name
fig_xlabels_KEYS = []
fig_xticks_KEYS = []
fig_data_KEYS = []
xtick = -1
for idx, row in csv_hd.iterrows():
    scheme_fullname = str(int(row[0]))
    if any(scheme in scheme_fullname for scheme in scheme_list):
        row_value = []
        for col in row.keys():
            if any(workload in col for workload in workload_dict):
                row_value.append(row[col])
        fig_data_KEYS.append(row_value)
        fig_xlabels_KEYS.append(scheme_fullname)
        xtick += 1
    fig_xticks_KEYS.append(xtick)
fig_data_KEYS = np.transpose(fig_data_KEYS)
assert fig_data_KEYS.shape[1] == len(fig_xlabels_KEYS) == len(fig_xticks_KEYS)

pprint(fig_data_KEYS)
```

```python
''' Ploting script '''
pp, fig = easypyplot.pdf.plot_setup(fig_path + fig_name, fig_dims)
# ax = fig.gca() # Only used with no subplot mode

''' write ratio '''
fig_ylim = (20, 70) # TODO: y-axis range. Align with your data
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1], 10.0))
fig_ax1 = fig.add_subplot(1, 4, 1) # Knob: plot the first subgraph of a 1x1 grid, i.e., no subplot mode
easypyplot.format.turn_off_box(fig_ax1)

fig_h = []
for group_id in range(len(workload_dict)):
    fig_h += fig_ax1.plot(fig_xticks_WR, fig_data_WR[group_id],
                marker=fig_marker[group_id],
                linewidth = '1.0',
                color=fig_colors[group_id],
                )

''' x axis '''
fig_ax1.xaxis.set_tick_params(pad=0) # Knob: distance between x tick label and x-axis
fig_ax1.set_xlim([fig_ax1.get_xticks()[0] - 0.5, fig_ax1.get_xticks()[-1] + 0.5])
fig_ax1.set_xticks(fig_xticks_WR)
fig_ax1.set_xticklabels(fig_xlabels_WR, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax1.xaxis.set_ticks_position('none') # Knob: uncomment for visible x ticks
fig_ax1.set_xlabel('write ratio', fontproperties=ft)
# ax.tick_params(direction='in') # Knob: [in, out, inout]
# ax.set_xlabel(fig_title) # Knob: the overall x label text

''' y axis '''
fig_ax1.yaxis.set_tick_params(pad=0) # Knob: distance between y tick label and y-axis
# easypyplot.format.set_axis_to_percent(fig_ax1.yaxis) # Uncomment for y axis tick label in xx% format
fig_ax1.yaxis.grid(True)
fig_ax1.set_ylim(fig_ylim)
fig_ax1.set_yticks(fig_yticks)
fig_yticks_label = ["{:.2f}".format(item) for item in fig_yticks] # Change for customized y tick label text
fig_ax1.set_yticklabels(fig_yticks_label, fontproperties=ft)
fig_ax1.set_ylabel(fig_title, multialignment='center', fontproperties=ft)
fig_ax1.tick_params(labelsize=xy_label_size)


''' util '''
fig_ylim = (10, 50) # TODO: y-axis range. Align with your data
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1], 10.0))
fig_ax2 = fig.add_subplot(1, 4, 2) # Knob: plot the first subgraph of a 1x1 grid, i.e., no subplot mode
easypyplot.format.turn_off_box(fig_ax2)

fig_h = []
for group_id in range(len(workload_dict)):
    fig_h += fig_ax2.plot(fig_xticks_U, fig_data_U[group_id],
                marker=fig_marker[group_id],
                linewidth = '1.0',
                color=fig_colors[group_id],
                )

''' x axis '''
fig_ax2.xaxis.set_tick_params(pad=0) # Knob: distance between x tick label and x-axis
fig_ax2.set_xlim([fig_ax2.get_xticks()[0] - 0.5, fig_ax2.get_xticks()[-1] + 0.5])
fig_ax2.set_xticks(fig_xticks_U)
fig_ax2.set_xticklabels(fig_xlabels_U, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax2.xaxis.set_ticks_position('none') # Knob: uncomment for visible x ticks
fig_ax2.set_xlabel('space utilization', fontproperties=ft)
# ax.tick_params(direction='in') # Knob: [in, out, inout]
# ax.set_xlabel(fig_title) # Knob: the overall x label text

''' y axis '''
fig_ax2.yaxis.set_tick_params(pad=0) # Knob: distance between y tick label and y-axis
# easypyplot.format.set_axis_to_percent(fig_ax2.yaxis) # Uncomment for y axis tick label in xx% format
fig_ax2.yaxis.grid(True)
fig_ax2.set_ylim(fig_ylim)
fig_ax2.set_yticks(fig_yticks)
fig_yticks_label = ["{:.2f}".format(item) for item in fig_yticks] # Change for customized y tick label text
fig_ax2.set_yticklabels(fig_yticks_label, fontproperties=ft)
# fig_ax2.set_ylabel(fig_title, multialignment='center', fontproperties=ft)
fig_ax2.tick_params(labelsize=xy_label_size)

''' KV size '''
fig_ylim = (0, 50) # TODO: y-axis range. Align with your data
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1], 10.0))
fig_ax3 = fig.add_subplot(1, 4, 3) # Knob: plot the first subgraph of a 1x1 grid, i.e., no subplot mode
easypyplot.format.turn_off_box(fig_ax3)

fig_h = []
for group_id in range(len(workload_dict)):
    fig_h += fig_ax3.plot(fig_xticks_KVS, fig_data_KVS[group_id],
                marker=fig_marker[group_id],
                linewidth = '1.0',
                color=fig_colors[group_id],
                )

''' x axis '''
fig_ax3.xaxis.set_tick_params(pad=0) # Knob: distance between x tick label and x-axis
fig_ax3.set_xlim([fig_ax3.get_xticks()[0] - 0.5, fig_ax3.get_xticks()[-1] + 0.5])
fig_ax3.set_xticks(fig_xticks_KVS)
fig_ax3.set_xticklabels(fig_xlabels_KVS, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax3.xaxis.set_ticks_position('none') # Knob: uncomment for visible x ticks
fig_ax3.set_xlabel('value size', fontproperties=ft)
# ax.tick_params(direction='in') # Knob: [in, out, inout]
# ax.set_xlabel(fig_title) # Knob: the overall x label text

''' y axis '''
fig_ax3.yaxis.set_tick_params(pad=0) # Knob: distance between y tick label and y-axis
# easypyplot.format.set_axis_to_percent(fig_ax3.yaxis) # Uncomment for y axis tick label in xx% format
fig_ax3.yaxis.grid(True)
fig_ax3.set_ylim(fig_ylim)
fig_ax3.set_yticks(fig_yticks)
fig_yticks_label = ["{:.2f}".format(item) for item in fig_yticks] # Change for customized y tick label text
fig_ax3.set_yticklabels(fig_yticks_label, fontproperties=ft)
# fig_ax3.set_ylabel(fig_title, multialignment='center', fontproperties=ft)
fig_ax3.tick_params(labelsize=xy_label_size)

''' key space '''
fig_ylim = (20, 50) # TODO: y-axis range. Align with your data
fig_yticks = list(np.arange(fig_ylim[0], fig_ylim[1], 5.0))
fig_ax4 = fig.add_subplot(1, 4, 4) # Knob: plot the first subgraph of a 1x1 grid, i.e., no subplot mode
easypyplot.format.turn_off_box(fig_ax4)

fig_h = []
for group_id in range(len(workload_dict)):
    fig_h += fig_ax4.plot(fig_xticks_KEYS, fig_data_KEYS[group_id],
                marker=fig_marker[group_id],
                linewidth = '1.0',
                color=fig_colors[group_id],
                )

''' x axis '''
fig_ax4.xaxis.set_tick_params(pad=0) # Knob: distance between x tick label and x-axis
fig_ax4.set_xlim([fig_ax4.get_xticks()[0] - 0.5, fig_ax4.get_xticks()[-1] + 0.5])
fig_ax4.set_xticks(fig_xticks_KEYS)
fig_ax4.set_xticklabels(fig_xlabels_KEYS, fontsize=fig_label_fontsize, fontproperties=ft)
fig_ax4.xaxis.set_ticks_position('none') # Knob: uncomment for visible x ticks
fig_ax4.set_xlabel('Key space', fontproperties=ft)
# ax.tick_params(direction='in') # Knob: [in, out, inout]
# ax.set_xlabel(fig_title) # Knob: the overall x label text

''' y axis '''
fig_ax4.yaxis.set_tick_params(pad=0) # Knob: distance between y tick label and y-axis
# easypyplot.format.set_axis_to_percent(fig_ax4.yaxis) # Uncomment for y axis tick label in xx% format
fig_ax4.yaxis.grid(True)
fig_ax4.set_ylim(fig_ylim)
fig_ax4.set_yticks(fig_yticks)
fig_yticks_label = ["{:.2f}".format(item) for item in fig_yticks] # Change for customized y tick label text
fig_ax4.set_yticklabels(fig_yticks_label, fontproperties=ft)
# fig_ax4.set_ylabel(fig_title, multialignment='center', fontproperties=ft)
fig_ax4.tick_params(labelsize=xy_label_size)

''' legend '''
legend = [workload_dict[item] for item in workload_dict]
fig.legend(fig_h, legend, loc='upper center', frameon=False, prop=ft,
        bbox_to_anchor=(0.5, 1.11),
        ncol=2,
        numpoints=1,
        columnspacing=5, # padding between columns
        labelspacing=1, # padding between rows
        )

plt.tight_layout()
# easypyplot.format.resize_ax_box(fig_ax1, hratio=0.8)
# easypyplot.format.resize_ax_box(fig_ax2, hratio=0.8)
# easypyplot.format.resize_ax_box(fig_ax3, hratio=0.8)
# easypyplot.format.resize_ax_box(fig_ax4, hratio=0.8)
```
